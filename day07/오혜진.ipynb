1. P217 스택 Code06-08.py의 선언된 함수를 자신이 좋아하는 형태의 이름으로 바꾸어 완성하시오. 실행결과를 P219와 다르게 테스트 해보세요.
예) isStackFull -> 스택풀확인 / amIfull 등...

```python
def checkStaFull():
    global SIZE, stack, top
    if top == (SIZE - 1) :
        return True 
    else : 
        return False
    
def checkStaEmpty() :
    global SIZE, stack, top
    if top <= -1 :
        return True # 스택이 비었음
    else : 
        return False  
    
def checkpush(data) : 
    global SIZE, stack, top
    if checkStaFull() == True : 
        print('stack is Full')
        return
    else : 
        top += 1 
        stack[top] = data 

def checkpop() : 
    global SIZE, stack, top
    if checkStaEmpty() == True : 
        return None
    else : 
        data = stack[top]
        stack[top] = None
        top -= 1
        return data

def checkpeek():
    global SIZE, stack, top 
    if checkStaEmpty() == True : 
        print('스택이 비어있습니다')
        return None
    else: 
        return stack[top]
    
SIZE = int(input('스택 크기를 입력하세요 --> '))
stack = [None for _ in range(SIZE)]
top = - 1 

if __name__ == '__main__' : 
    while True : 
        select = input('PUSH(p),POP(o),PEEK(e),EXIT(x)>' )

        if select.lower() == 'p': 
            data = input('입력데이터 >> ')
            checkpush(data)
            print(f'스택상태 : {stack}')
        elif select.lower() == 'o': 
            data = checkpop()
            print(f'추출데이터 : {data}')
            print(f'스택상태 : {stack}')
        elif select.lower() == 'e': 
            data = checkpeek()
            print(f'확인데이터 : {data}')
            print(f'스택상태 : {data}')
        elif select.lower() == 'x' : 
            break
        else : 
            continue
```

2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
def isQueueFull() :
	global SIZE, queue, front, rear
	if (rear == SIZE-1) :
		return True
	else :
		return False

def isQueueEmpty() :
	global SIZE, queue, front, rear
	if (front == rear) :
		return True
	else :
		return False

def enQueue(data) :
	global SIZE, queue, front, rear
	if (isQueueFull()) :
		print("큐가 꽉 찼습니다.")
		return
	rear += 1
	queue[rear] = data

def deQueue() :
	global SIZE, queue, front, rear
	if (isQueueEmpty()) :
		print("큐가 비었습니다.")
		return None
	front += 1
	data = queue[front]
	queue[front] = None

	for i in range(front + 1, rear + 1): 	
		queue[i - 1] = queue[i]
		queue[i] = None
	front = -1
	rear -= 1

	return data

def peek() :
	global SIZE, queue, front, rear
	if (isQueueEmpty()) :
		print("큐가 비었습니다.")
		return None
	return queue[front+1]

## 전역 변수 선언 부분 ##
SIZE = 5
queue = [ None for _ in range(SIZE) ]
front = rear = -1

## 메인 코드 부분 ##
if __name__ == "__main__":
    while True:
        action = input('손님대기(p), 손님입장(o), 다음입장손님확인(s), 종료(q): ')

        if action == "p": 
            data = input("대기할 손님의 이름을 입력하세요: ")
            enQueue(data)
            print("대기 줄 상태:", queue)
        elif action == "o":  
            if isQueueEmpty():
                print("대기 중인 손님이 없습니다.")
            else:
                print(deQueue(), "님이 식당에 들어감")
                print("대기 줄 상태:", queue)
        elif action == "s": 
            next_guest = peek()
            if next_guest:
                print("다음 입장 손님은", next_guest, "입니다.")
            else:
                print("대기 중인 손님이 없습니다.")
        elif action == "q":  
            print("식당 영업 종료!")
            break
        else:
            print("올바른 명령을 입력하세요.")
```

3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.

```python
import random
## 함수 선언 부분 ##
class TreeNode() :
	def __init__ (self) :
		self.left = None
		self.data = None
		self.right = None

## 전역 변수 선언 부분 ##
memory = []
root = None

print("오늘 판매된 물건을 입력하세요. 입력을 마치려면 '입력완료'을 입력하세요.")

dataAry = []
while True:
    item = input("물건 이름을 입력하세요: ")
    if item == "입력완료":
        break
    else:
        dataAry.append(item), print("입력된 물건:", item)


sellAry = [random.choice(dataAry) for _ in range(20)]

print('오늘 판매된 물건(중복O) -->', sellAry)

## 메인 코드 부분 ##
node = TreeNode()
node.data = sellAry[0]
root = node
memory.append(node)

for name in sellAry[1:] :

	node = TreeNode()
	node.data = name

	current = root
	while True :
		if name == current.data :
			break
		if name < current.data :
			if current.left == None :
				current.left = node
				memory.append(node)
				break
			current = current.left
		else :
			if current.right == None :
				current.right = node
				memory.append(node)
				break
			current = current.right

print("이진 탐색 트리 구성 완료!")

def preorder(node) :
	if node == None :
		return
	print(node.data, end = ' ')
	preorder(node.left)
	preorder(node.right)

print('오늘 판매된 종류(중복X)--> ', end = ' ')
preorder(root)
```
